```python
from turfpy.measurement import boolean_point_in_polygon
from geojson import Point, Polygon, Feature
def calc_totalrainfall(df,rd):
  lat_in=df['LATITUDE_IB_CW']
  lat_out=df['LATITUDE_OB_CCW']
  lon_in=df['LONGITUDE_IB_CW']
  lon_out=df['LONGITUDE_OB_CCW']
  #frdf=rd[(rd['LAT']>=lat_in) & (rd['LAT']<=lat_out) & (rd['LON']>=lon_in) & (rd['LON']<=lon_out)]
  lon_lst=[lon_in-0.05,lon_out+0.05,lon_in-0.05,lon_out+0.05]
  latlst=[lat_in-0.05, lat_in-0.05, lat_out+0.05, lat_out+0.05]
  polygon_geom = Polygon([list(zip(lon_lst, latlst))])
  
  rd['status']=rd.apply(geopt_inside_polygon,polygon_g=polygon_geom,axis=1)
  filt_df=rd[rd['status']==True]
  dt=df['DETECTION_DATE_TIME']
  prevday=dt - timedelta(days=1)
  tdstr=dt.strftime('%d-%b')
  prevdstr=prevday.strftime('%d-%b')
  if prevdstr in rd.columns:
    
    total_rain=filt_df[tdstr].sum()+filt_df[prevdstr].sum()
  else: 
    total_rain=filt_df[tdstr].sum()
  return total_rain
```

```python
hdf_dum['toatl_rain']=hdf_dum.apply(calc_totalrainfall,rd=rdf,axis=1)
```
