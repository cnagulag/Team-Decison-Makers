```python
def getGeodesicDistInMiles(coords1, coords2):
    # Importing the geodesic module from the library
    from geopy.distance import geodesic
    # Print the distance calculated in km
    return((geodesic(coords1, coords2).km)*0.62137)
```

```python
def getGeodesicDistInMeters(coords1, coords2):
    # Importing the geodesic module from the library
    from geopy.distance import geodesic
    # Print the distance calculated in km
    return((geodesic(coords1, coords2).km)*1000)
```

```python
def getRainfall(df, rf):
    rain_totals =[]
    rain_dist = []
    dist = {}
    
    for index, rows in df.iterrows():
        coords1 = [rows[0], rows[1]]
        # define a dict for distances
        dist = {}
        for ind in rf.index:
            coords2 = [rf['LAT'][ind], rf['LON'][ind]]
            dist.update({ind : getGeodesicDistInMeters(coords1, coords2)})
        min_ind = getIndexOfMin(dist)
        rain_totals.append(rf['Totals'][min_ind])
        rain_dist.append(min(dist.values()))
    
    df['Rain_Totals'] = rain_totals;
    df['Rain_Dist'] = rain_dist;
    print("Successfully Finished calculating the rainfall")
```

```python
def getIndexOfMin(dist):
    min_dist = min(dist.values());
    idx = 0;
    for key, value in dist.items():
        if value == min_dist:
            idx = key;
    return idx;
```
