```python
import pandas as pd;
```

```python
from geopy.geocoders import Nominatim
```

```python

# initialize Nominatim API 
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
def getCounty(data, lat_column, lon_column):
    counties =[]
    for lat, lon in zip(data[lat_column], data[lon_column]):
        location = geolocator.reverse(str(lat)+","+str(lon))
        add = location.raw['address']
        counties.append(add['county'])
    
    data['county_calc'] = counties;

#location = geolocator.reverse(Latitude+","+Longitude)
#add = location.raw['address']
#print(add['county'])
```

```python
df = pd.read_csv("harvey_highwater_incidents_RECENT.csv")
rf = pd.read_csv("Rainfall-V5.csv")
```
