```python
coord_list_dict = []
for ind, row in geo_df.iterrows():
    coord_list_dict.append({'coordinates' : [[row['LONGITUDE_IB_CW'], row['LATITUDE_IB_CW']], [row['LONGITUDE_OB_CCW'], row['LATITUDE_OB_CCW']]],
                           'dates': [str(row['ENTER_DT']), str(row['CLEAR_DT'])],
                           'color': colors.get(row['COUNTY_NAME']),
                           'weight': int(row['elev_meters']),
                           'popup': ("Incident:"+ str(ind) + "\n Elevation: " + str(round(row['elev_meters'])) + "\nRainfall:" + str(round(row['Rain_Totals_meters'], 3)) + 
          "\nROAD:" + row['ROADWAY_NAME'])})
```

```python
path = pd.read_csv("Harvey_Path_V1.csv")
```

```python
crs = {'init': 'epsg:4326'}
geom = [Point(xy) for xy in zip(path["LON"], path["LAT"])]
geom[:3]
geo_path = gpd.GeoDataFrame(path, crs = crs, geometry = geom)
geo_path.head()
```

```python
m = folium.Map(location = [30, -100], tiles='Stamen Terrain', zoom_start = 5.0, style = "dark_only_labels")
features = [
    {
        "type": "Feature",
        "geometry": {
            "type": "LineString",
            "coordinates": line["coordinates"],
        },
        "properties": {
            "times": line["dates"],
            "style": {
                #"color": line["color"],
                #"weight": line["weight"] if "weight" in line else 5,
                'weight' :1
            },
            "popup": line["popup"],
            'icon': 'circle',
                'iconstyle':{
                    'color': line["color"],
                    'fillOpacity': 0.2,
                    'fillcolor':'crimson',
                    'stroke': 'true',
                    #'weight': line["weight"],
                    'radius': line["weight"]/2
                },
        },
    }
    for line in coord_list_dict
]

plugins.TimestampedGeoJson(
    {
        "type": "FeatureCollection",
        "features": features,
    },
    period= "PT1H", #"PT1M",
    add_last_point=True,
    max_speed=10000
).add_to(m)

paths = []
paths = [[point.xy[1][0], point.xy[0][0]] for point in geo_path.geometry]
plugins.MarkerCluster(paths).add_to(m)
m
#Make this Notebook Trusted to load map: File -> Trust Notebook
m.save('map2.html')
```
